```python
#Rounding the values in the dataset to 2 decimal places
round(df_sales,2)
```

```python
# Spliting the date column to day, month and year
df_sales["Day"] = (df_sales["Date"]).dt.day
df_sales["Month"] = (df_sales["Date"]).dt.month
df_sales["Year"] = (df_sales["Date"]).dt.year
```

```python
# splitting the time column to Hours
df_sales['Time'] = pd.to_datetime(df_sales['Time'], format='%H:%M:%S')
df_sales['Hour'] = (df_sales["Time"]).dt.hour
```

## Gender Analysis

### Gender distribution

```python
df_sales.value_counts("Gender")
```

```python
fig, ax = plt.subplots(figsize=(4,4))
sns.countplot(x="Gender",data=df_sales).set_title("Gender Distribution")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x(), p.get_height() +1))
```

#### From the data the difference between the gender distribution is very small, however the Females appear to do more of the shopping with  502 females recorded and 409 males recorded 

### Sales by Gender

```python
df_sales.pivot_table (index="Gender",
                     values=["Quantity","Total"],
                     aggfunc="sum").sort_values("Total",ascending=False)
```

```python
fig,ax=plt.subplots(figsize=(3,4))
sns.set(style="darkgrid")
sns.barplot(x="Gender",y="Total",data=df_sales, estimator=sum)
plt.title("Sales by Gender")
for p in ax.patches:
    value = round(p.get_height(), 2)
    label = "{:.2f}".format(value)
    ax.annotate(label, (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
df_sales.groupby(["Gender"]).sum().plot(kind="pie",x="Gender",y="Total",autopct="%1.1f%%",startangle=90)
plt.legend(loc="lower right",fontsize=8)
plt.title("Percentage Contribution of Gender to Sales")
```

#### The most contribution to sales is from the female gender with a total of 167,882.92 revenue generated from 2,869 ordered quantites, this a makes up 52% of the total revenue generated  

### Average Rating by Gender

```python
df_sales.groupby(["Gender"]).agg({"Rating":"mean"}).sort_values("Rating",ascending=False)
```

```python
plt.figure(figsize=(3,4))
sns.boxplot(x="Gender",y="Rating",data=df_sales).set_title("Average Rating by Gender")

plt.show()
```

#### On the average, both genders gave the same rating of about 6.9
